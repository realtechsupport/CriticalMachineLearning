```python
import pandas
#inputfile = 'BNC2_sample.csv'
#source = datapath + inputfile
source = 'https://raw.githubusercontent.com/realtechsupport/CriticalMachineLearning/main/various_datasets/BNC2_sample.csv'
# Load data as a dataframe
#dataframe = pandas.read_csv(source, sep = ',')

# Read BNC2 sample dataset
dataframe = pandas.read_csv(source, names=['Code', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'VWAP', 'TWAP'])

# Show first 5 rows
dataframe.head(5)
```

```python
coins = dataframe.Code.unique()
coin1 = coins[0]
coin2 = coins[10]
selected_coins = [coin1, coin2]
targetdate = '2018-01-01'
dataframe[dataframe.Code.isin(selected_coins) & (dataframe.Date == targetdate)]
```

```python
# filter unwanted entries
# Number of observations in dataset
print( 'Before:', len(dataframe) )

# Get all the GWA codes
gwa_codes = [code for code in dataframe.Code.unique() if 'GWA_' in code]

# Only keep GWA observations
dataframe = dataframe[dataframe.Code.isin(gwa_codes)]

# Number of observations left
print( 'After:', len(dataframe) )
```

```python
#Pivot the dataframe
pivoted_df = dataframe.pivot(index='Date', columns='Code', values='VWAP')

# Display examples from pivoted dataset
pivoted_df.tail()
```

```python
# Shift by some period

print(pivoted_df.tail(3))
print()
print(pivoted_df.tail(3).shift(1))
```

```python
# Calculate returns over 7 days
delta_7 = pivoted_df / pivoted_df.shift(7) - 1.0

# Display examples
delta_7.tail()
```

```python
# Calculate returns over each window and store them in dictionary
delta_dict = {}
for offset in [7, 14, 21, 28]:
    delta_dict['delta_{}'.format(offset)] = pivoted_df / pivoted_df.shift(offset) - 1.0
```

```python
import itertools
# Using islice() + items() Get first N items in dictionary
N = 2
out = dict(itertools.islice(delta_dict.items(), N))
print(str(out))
```
