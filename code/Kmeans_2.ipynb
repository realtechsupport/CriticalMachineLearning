```python
# Demonstrate kmeans on a classic dataset (IRIS)
# https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html
# Updated FEB 2024

import matplotlib.pyplot as plt
import pandas
import numpy
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
# Load Sklearn IRIS dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
#Show the data in a loop
N = 5
for i in range(0, N):
  print(X[i])
  print(y[i])
```

```python
# Use pandas to display the data
pd_data = pandas.DataFrame(data=iris.data, columns=iris.feature_names)
pd_data.head(200)
```

```python
# Create an instance of K-Means
n = 3
kmc = KMeans(n_clusters=n, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)

# Fit and make predictions
# the function returns the array of cluster labels earch dat point belongs to
y_kmc = kmc.fit_predict(X)
```

```python
# Check out the results

#What are the labels for each point
print(kmc.labels_)

# How many unique values (clusters) are there
print(numpy.unique(kmc))

# What are the cluster centers
print(kmc.cluster_centers_)
```

```python
# Show the original data
plt.scatter(X[:, 0], X[:, 1], c=y_kmc, s=50, cmap='viridis')
# Show the clusters
centers = kmc.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.5, label='centroids');

plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Show the original data
plt.scatter(X[:, 0], X[:, 3], c=y_kmc, s=50, cmap='viridis')
# Show the clusters
centers = kmc.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 3], c='red', s=100, alpha=0.5, label='centroids');

plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Show the original data
plt.scatter(X[:, 2], X[:, 3], c=y_kmc, s=50, cmap='viridis')
# Show the clusters
centers = kmc.cluster_centers_
plt.scatter(centers[:, 2], centers[:, 3], c='red', s=100, alpha=0.5, label='centroids');

plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Now predict which class a new data set would belong to
# sepal length (cm)	sepal width (cm)	petal length (cm)	petal width (cm)
example_1 = [[5.0,	3.2, 1.3,	0.2]]
result = kmc.predict(example_1)
print(result)

example_2 = [[6.3,	3.4,	5.5,	2.4]]
result = kmc.predict(example_2)
print(result)
```
