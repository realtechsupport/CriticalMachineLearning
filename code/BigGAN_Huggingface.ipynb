```python
from diffusers import DiffusionPipeline
```

```python
# pick a model from the collection: https://huggingface.co/models?library=diffusers
amodel = "CompVis/ldm-text2im-large-256"
#call only once (if you don't want to change models)
generator = DiffusionPipeline.from_pretrained(amodel)
```

```python
prompt1 = "a garden on a rainy day"
prompt2 = "a garden on a sunny day"
image1 = generator(prompt1).images[0]
image2 = generator(prompt2).images[0]

rows = 1
columns = 2
fig = plt.figure(figsize=(16,16), dpi=80)

fig.add_subplot(rows, columns, 1)
plt.imshow(image1)

fig.add_subplot(rows, columns, 2)
plt.imshow(image2)
```

```python
prompt1 = "a garden on a sunny day with peach trees, apple trees and pear trees."
prompt2 = "a garden on a sunny day with dogs, cats, squirrels and rabbits."
image1 = generator(prompt1).images[0]
image2 = generator(prompt2).images[0]

rows = 1
columns = 2
fig = plt.figure(figsize=(16,16), dpi=80)

fig.add_subplot(rows, columns, 1)
plt.imshow(image1)

fig.add_subplot(rows, columns, 2)
plt.imshow(image2)
```
