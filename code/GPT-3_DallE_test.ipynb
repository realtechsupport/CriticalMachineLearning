```
import openai
```

```
#prompt = sys.argv[1]
#openai.api_key = os.environ['api_key']
openai.api_key = APIkey
#this_engine = "text-davinci-003"
this_engine = "gpt-3.5-turbo"
tokens = 4096
temp = 0.75

prompt = "Why was the Industrail revolution important in the UK"
ntokens = tokens -len(prompt)

if(this_engine == "text-davinci-003"):
  completions = openai.Completion.create(engine=this_engine, prompt=prompt, max_tokens=ntokens, n=1, stop=None, temperature=temp)
  message = completions.choices[0].text

elif(this_engine == "gpt-3.5-turbo"):
  completion = openai.ChatCompletion.create(model = 'gpt-3.5-turbo', messages = [{'role': 'user', 'content': prompt}],temperature=temp)
  message = completion['choices'][0]['message']['content']
 
print(message)
```

```
import openai
openai.api_key = APIkey
num = 1
asize="1024x1024"
bsize = "256x256"
idea = "hot air balloon playing cards with a fox wearing a red hat"
```

```
#If you want to look at the resultant image
response = openai.Image.create(prompt=idea, n=1,size=bsize)
image_url = response['data'][0]['url']
#(click on the url)
print(image_url)
```

```
# Generic image storage.pip
image_name = "dalle_image.jpg"
data = requests.get(image_url).content
f = open(image_name,'wb')
f.write(data)
f.close()
```

```
# If you want to save the image on Colab Ddrive ..
# get the returned object as ajson object and save the image to the datapath
# https://realpython.com/generate-images-with-dalle-openai-api/

import json
from base64 import b64decode
from pathlib import Path

response = openai.Image.create(prompt=idea, n=1,size=bsize, response_format="b64_json")
file_name = datapath +  f"{idea[:5]}-{response['created']}.json"
with open(file_name, mode="w", encoding="utf-8") as file:
    json.dump(response, file)

with open(file_name, mode="r", encoding="utf-8") as file:
    response = json.load(file)

for index, image_dict in enumerate(response["data"]):
  image_data = b64decode(image_dict["b64_json"])
  image_file = file_name.split('.json')[0] + ".jpg"
  with open(image_file, mode="wb") as jpg:
    jpg.write(image_data)


```
